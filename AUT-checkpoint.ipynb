```python
from datetime import datetime

# Carregar o DataFrame a partir do arquivo Excel filtrado
tabela = pd.read_excel("tabela_filtrada.xlsx")

# Converter os valores da coluna para minúsculas
tabela['A pessoa dentro do Espectro Autista tem plano de Saúde?'] = tabela['A pessoa dentro do Espectro Autista tem plano de Saúde?'].str.lower()

# Filtrar apenas as pessoas que não têm plano de saúde
pessoas_sem_plano = tabela[tabela['A pessoa dentro do Espectro Autista tem plano de Saúde?'] == 'não'].copy()

# Verificar se a coluna 'Data de Nascimento ' está presente
if 'Data de Nascimento ' in pessoas_sem_plano.columns:
    # Certificar-se de que a coluna 'Data de Nascimento ' está no formato de data
    pessoas_sem_plano['Data de Nascimento '] = pd.to_datetime(pessoas_sem_plano['Data de Nascimento '], errors='coerce')
    
    # Calcular a idade
    data_atual = datetime.now()
    pessoas_sem_plano['Idade'] = pessoas_sem_plano['Data de Nascimento '].apply(lambda x: data_atual.year - x.year - ((data_atual.month, data_atual.day) < (x.month, x.day)) if pd.notnull(x) else None)
else:
    print("A coluna 'Data de Nascimento ' não está presente no DataFrame.")

# Selecionar as colunas desejadas
colunas_desejadas = ['Nome da Mãe, Pai ou Responsável', 'Nome da pessoa dentro Espectro Autista (TEA)', 'Idade']
nova_tabela_sem_plano = pessoas_sem_plano[colunas_desejadas]

# Exibir a nova tabela
print("\nTabela de pessoas sem plano de saúde:")
print(nova_tabela_sem_plano)
display(nova_tabela_sem_plano)
# Salvar a nova tabela em um novo arquivo Excel
nova_tabela_sem_plano.to_excel("tabela_sem_plano_com_idades.xlsx", index=False)
```

```python
import os

def substituir_emails_arquivo(caminho_arquivo, novo_caminho_arquivo):
    # Abrir o arquivo original para leitura
    with open(caminho_arquivo, 'r', encoding='utf-8') as arquivo_orig:
        # Ler o conteúdo do arquivo
        conteudo = arquivo_orig.read()
        # Substituir todos os endereços de e-mail por um marcador
        novo_conteudo = conteudo.replace('@', 'EMAIL_MASCARADO')

    # Escrever o conteúdo modificado em um novo arquivo
    with open(novo_caminho_arquivo, 'w', encoding='utf-8') as novo_arquivo:
        novo_arquivo.write(novo_conteudo)

def substituir_emails_planilha(caminho_planilha, novo_caminho_planilha):
    # Carregar a planilha
    tabela = pd.read_excel(caminho_planilha)

    # Substituir todos os endereços de e-mail na planilha por um marcador
    tabela_sem_email = tabela.applymap(lambda x: 'EMAIL_MASCARADO' if isinstance(x, str) and '@' in x else x)

    # Salvar a planilha com os e-mails substituídos em um novo arquivo
    tabela_sem_email.to_excel(novo_caminho_planilha, index=False)

# Definir o caminho para seus arquivos de código
caminho_para_arquivos_de_codigo = 'C:\\Users\\Dell\\Desktop\\aut\\JupyterAUT'

# Lista todos os arquivos no diretório
arquivos = os.listdir(caminho_para_arquivos_de_codigo)

# Iterar sobre cada arquivo
for arquivo in arquivos:
    # Verificar se é um arquivo Python
    if arquivo.endswith('.py'):
        # Substituir e-mails no arquivo
        substituir_emails_arquivo(os.path.join(caminho_para_arquivos_de_codigo, arquivo), 
                                   os.path.join(caminho_para_arquivos_de_codigo, 'temp_' + arquivo))
        # Remover o arquivo original
        os.remove(os.path.join(caminho_para_arquivos_de_codigo, arquivo))
        # Renomear o arquivo temporário para o nome original
        os.rename(os.path.join(caminho_para_arquivos_de_codigo, 'temp_' + arquivo), 
                  os.path.join(caminho_para_arquivos_de_codigo, arquivo))

# Substituir e-mails na planilha Excel
substituir_emails_planilha(os.path.join(caminho_para_arquivos_de_codigo, 'aut.xlsx'), 
                           os.path.join(caminho_para_arquivos_de_codigo, 'temp_aut.xlsx'))
# Remover a planilha original
os.remove(os.path.join(caminho_para_arquivos_de_codigo, 'aut.xlsx'))
# Renomear a planilha temporária para o nome original
os.rename(os.path.join(caminho_para_arquivos_de_codigo, 'temp_aut.xlsx'), 
          os.path.join(caminho_para_arquivos_de_codigo, 'aut.xlsx'))

print("Substituição de e-mails concluída com sucesso!")
```
